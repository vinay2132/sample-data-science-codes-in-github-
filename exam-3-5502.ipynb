```python
myDict = {'love': 0.9, 'awesome': 0.8, 'cool': 0.5, 'good': 0.4, 'okay': 0.2, 'bad': -0.4, 'terrible': -0.6, 'disgusting': -0.7, 'hate': -0.9}

def main_sentiment_analysis(comment):
    words = comment.split()
    total_happiness = 0
    for word in words:
        if word in myDict:
            total_happiness += myDict[word]
    avg_happiness = total_happiness / len(words)
    return avg_happiness
```

```python
comment = "I love this product! It's awesome!"
print(main_sentiment_analysis(comment))
```

```python
wordlist = comment.split(" ")
print(wordlist)
print(main_sentiment_analysis(comment))
```
