```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
from sklearn import tree

music_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/music.csv")
music_data
```

Preparing the data :
cleaning and removing dublicate values but the data doen't have but we need to divide the data into two parts input data(age and gender) and output(genre)

```python
x = music_data.drop(columns=['genre'])
y = music_data['genre']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

Learing and predection:Algorithem and we are using decision tree we use DecisionTreeClassifier

Calculating the accuracy to make the model perfect and measur the accuracy
so to do this web need to split the data set into training and testing data set we use train_test_split for this

we need to have good amount of data for training and testing to make the model accurate and this can be calcuated by accuracy_score


```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
#predction = model.predict([ [21,1],[22,0] ])
predction = model.predict(x_test)
#predction
score = accuracy_score(y_test,predction)
score
```
