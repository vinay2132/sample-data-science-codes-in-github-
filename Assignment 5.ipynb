# Part 1: Data Visualizations
1. Grouped Bar Plots:


Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

Children = [780, 1050, 3056, 5025]
Adults = [315, 400, 1000, 1500]

br1 = np.arange(len(Children))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, Children, color ='r', width = barWidth,
		edgecolor ='grey', label ='Children')
plt.bar(br2, Adults, color ='g', width = barWidth,
		edgecolor ='grey', label ='Adult')

plt.xlabel('Months', fontweight ='bold', fontsize = 15)
plt.ylabel('Total no of Adults and Children’s ', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Children))],
		['April', 'May', 'June', 'July'])

plt.legend()
plt.show()
```

Stacked Bar Plot 

```python
import matplotlib.pyplot as plt
 

barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

x = ['April', 'May', 'June', 'July']
Children = [780, 1050, 3056, 5025]
Adult = [315, 400, 1000, 1500]
 
plt.xlabel('Months', fontweight ='bold', fontsize = 15)
plt.ylabel('Total no of Adults and Children’s ', fontweight ='bold', fontsize = 15)
plt.bar(x, Children, color='r')
plt.bar(x, Adults, bottom=Children, color='g')
plt.legend(["Children", "Adults"])
plt.show()
```

```python
counter = 1
for i in range(8):
    for j in range(8):
        myArray[i, j] = counter
        counter += 1

# print the resulting array
print(myArray)
```

# Part 3: Data Frames

1. Input the following data into a data frame called titanic, and display the entire data frame:

 Sex, Class, Survived, Died

 Children, First, 6, 0

 Children, Second, 24, 0

 Children, Third, 27, 52

 Men, First, 57, 118

 Men, Second, 14, 154

 Men, Third, 75, 387

 Men, Crew, 192, 693

 Women, First, 140, 4

 Women, Second, 80, 13

 Women, Third, 76, 89
 
 Women, Crew, 20, 3 


```python
import pandas as pd

# create a list of dictionaries with the data
data = [
    {'Sex': 'Children', 'Class': 'First', 'Survived': 6, 'Died': 0},
    {'Sex': 'Children', 'Class': 'Second', 'Survived': 24, 'Died': 0},
    {'Sex': 'Children', 'Class': 'Third', 'Survived': 27, 'Died': 52},
    {'Sex': 'Men', 'Class': 'First', 'Survived': 57, 'Died': 118},
    {'Sex': 'Men', 'Class': 'Second', 'Survived': 14, 'Died': 154},
    {'Sex': 'Men', 'Class': 'Third', 'Survived': 75, 'Died': 387},
    {'Sex': 'Men', 'Class': 'Crew', 'Survived': 192, 'Died': 693},
    {'Sex': 'Women', 'Class': 'First', 'Survived': 140, 'Died': 4},
    {'Sex': 'Women', 'Class': 'Second', 'Survived': 80, 'Died': 13},
    {'Sex': 'Women', 'Class': 'Third', 'Survived': 76, 'Died': 89},
    {'Sex': 'Women', 'Class': 'Crew', 'Survived': 20, 'Died': 3}
]

# create the dataframe
titanic = pd.DataFrame(data)

# display the dataframe
print(titanic)
```

2. Now only show the data of the people in first class.


```python
first_class = titanic[titanic['Class'] == 'First']
print(first_class)
```

3. Delete the crew members from the data.


```python
titanic.drop(titanic.index[(titanic['Class'] == 'Crew')],axis=0,inplace=True)
print(titanic)
```

4. Create a new column that is the total number of people for that group (those who survived + died).


```python
titanic['Total'] = titanic.apply(lambda row: row['Survived'] + row['Died'], axis=1)
print(titanic)
```

5. Create a new column with the percentage of people who survived.


```python
titanic['Survival %'] = titanic.apply(lambda row: row['Survived'] / (row['Survived'] + row['Died']) * 100, axis=1)
print(titanic)
```

6. Delete the column indicating the total number of people in that group.


```python
titanic.drop('Total', axis=1, inplace=True)
print(titanic)
```

7. Only show the rows where more than 80% of the people survived.

```python
survival_percentage = titanic['Survival %']
filtered_rows = titanic[survival_percentage > 80]
print(filtered_rows)
```

8. Then only show the rows where less than 40% of the people survived.


```python
survival_percentage = titanic['Survival %']
filtered_rows = titanic[survival_percentage < 40]
print(filtered_rows)
```

9. Calculate the total number of people that survived and died for each class, then report the percentages. (Hint: Use a grouped calculation.)


```python
grouped_data = titanic.groupby('Class').sum()
grouped_data['Total'] = grouped_data['Survived'] + grouped_data['Died']
grouped_data['Survival Percentage'] = grouped_data['Survived'] / grouped_data['Total'] * 100
print(grouped_data)
```
