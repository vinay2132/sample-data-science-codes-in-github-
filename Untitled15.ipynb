```python
main_model_proportions = make_array(0.01,0.99)

def simulate():
    return(50*sample_proportions(50,main_model_proportions).item(0))

simulate()
```

```python
visited_test_under_null = make_array()
repetitions = 20000

for i in np.arange(repetitions):
    new_num_of_correct_area_code = simulate()
    visited_test_statistics_under_null = np.append(visited_test_statistics_under_null,new_num_of_correct_area_code)
    
    
visited_test_statistics_under_null
```
