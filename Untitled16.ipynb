```python

Alternative Hypothesis: There is a higher likelihood that spammers are using the 940 area code in the population.
```

```python

model_proportions = make_array(0.01,0.99)

def simulate():
    return(50*sample_proportions(50,model_proportions).item(0))

simulate()
```

```python

test_statistics_under_null = make_array()
repetitions = 20000

for i in np.arange(repetitions):
    new_num_of_correct_area_code = simulate()
    test_statistics_under_null = np.append(test_statistics_under_null,new_num_of_correct_area_code)
    
    
test_statistics_under_null
```

```python
len(test_statistics_under_null) == 20000
```

```python
import matplotlib.pyplot as plt

x = test_statistics_under_null
num_bins = np.arange(0, 5, 1)

n, bins, patches = plt.hist(x, num_bins, facecolor = 'blue', alpha = 0.75, edgecolor = 'black')

plt.title('Hist')
plt.show()
```

```python
observed_statistic = np.sum(df['Area_Code'] == 940)
p_value = np.mean(test_statistics_under_null >= observed_statistic)
print("Empirical p-value:", p_value)
```
