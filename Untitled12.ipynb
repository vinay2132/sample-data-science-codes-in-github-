```python
import math

def standard_deviation(datalist):
    # Compute the mean of the data
    mean = sum(datalist) / len(datalist)
    
    # Compute the variance of the data
    variance = sum((x - mean) ** 2 for x in datalist) / (len(datalist) - 1)
    
    # Compute the standard deviation of the data
    standard_deviation = math.sqrt(variance)
    
    # Print the mean, variance and standard deviation of the data
    print("Mean: {:.2f}".format(mean))
    print("Variance: {:.2f}".format(variance))
    print("Standard Deviation: {:.2f}".format(standard_deviation))
```

```python
data = [1, 2, 3, 4, 5]
standard_deviation(data)
```
