removing string values


```
sex=pd.get_dummies(titanic_data['Sex'],drop_first=True)
sex.head(5)
```

```
embark=pd.get_dummies(titanic_data['Embarked'],drop_first=True)
embark
```

```
pcl=pd.get_dummies(titanic_data['Pclass'],drop_first=True)
pcl.head(5)
```

```
titanic_data = pd.concat([titanic_data,sex,embark,pcl],axis=1)
titanic_data
```

```
titanic_data.drop(['Sex',	'Embarked','PassengerId','Pclass','Name','Ticket'],axis=1,inplace=True)
```

```
titanic_data
```

spliting the data into train and test subset data


```
X=titanic_data.drop("Survived",axis=1)
y=titanic_data['Survived']
```

```
import sklearn as sk 
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```
from sklearn.linear_model import LogisticRegression
```

```
logmodal = LogisticRegression()
```

```
logmodal.fit(X_train,y_train)
```

```
prediction = logmodal.predict(X_test)
```

```
from sklearn.metrics import classification_report
```

```
classification_report(y_test,prediction)
```

```
from sklearn.metrics import confusion_matrix
```

```
confusion_matrix(y_test,prediction)
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y_test,prediction)
```
