```python
from numpy import char
import math
max_pickups = 3
edge_length = 1
clock_tick = 0

class Van:
    def __init__(self, id, location):
        self.id = id
        self.location = location

class Customer:
    def __init__(self, id, location):
        self.id = id
        self.location = location

def calculated_distance(node1,node2):
  return edge_length

def customer_pickup(customers_main, van_location):
  customers_main = sorted(customers_main, key=lambda x: calculated_distance(van_location, x[0]))
                   
  pickups = []
  for customer in customers_main:
    if len(pickups) == max_pickups:
      break
    pickups.append(customer)
  return pickups

def customer_drop_off(customers_main, van_location):
  customers_main = sorted(customers_main, key=lambda x: calculated_distance(van_location, x[1])) 
  return customers_main

def assign_customers_to_vans(customers, vans):
    assigned_vans = {van.id: [] for van in vans}
    for customer in customers:
        min_distance = math.inf
        assigned_van = None
        for van in vans:
            distance = calculate_distance(van.location, customer.location)
            if distance < min_distance:
                min_distance = distance
                assigned_van = van
        assigned_vans[assigned_van.id].append(customer)
        assigned_van.location = customer.location
    return assigned_vans

def plan_route_for_van(van, assigned_customers):
    drop_off_locations = [customer.location for customer in assigned_customers]
    drop_off_locations.sort(key=lambda loc: calculate_distance(van.location, loc))
    route = [van.location] + drop_off_locations
    van.location = drop_off_locations[-1]
    return route

def calculate_distance(loc1, loc2):
    return math.sqrt((loc1[0]-loc2[0])**2 + (loc1[1]-loc2[1])**2 + clock_tick)

if __name__ == '__main__':
    customers = [Customer(1, (0,0)), Customer(2, (3,4)), Customer(3, (5,1))]
    vans = [Van(1, (1,1)), Van(2, (4,5)), Van(3, (6,2))]

    customers_main = [(0,0), (3,4), (5,1)]
    van_location = 0
    pickup_orders = customer_pickup(customers_main, van_location)
    drop_off_route = customer_drop_off(pickup_orders, van_location)

    assigned_vans = assign_customers_to_vans(customers, vans)

    for van_id, assigned_customers in assigned_vans.items():
        van = next(van for van in vans if van.id == van_id)
        route = plan_route_for_van(van, assigned_customers)
        print(f"Van assigned to P{', '.join(str(customer.id) for customer in assigned_customers)}")
        print(f"Route: {', '.join(str(loc) for loc in route)}\n")
    print(f"Pickup orders:", pickup_orders)
    print(f"Drop off routes:", drop_off_route)
       
```
