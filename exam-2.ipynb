Step : 3 Step 3: Now, only using the following training_data and training_target data sets, create a survival classifier given someone’s class, age, and sex. Use a kNN classifier with k=1, then apply it to a simple test case of your choosing.

```python
from sklearn.neighbors import KNeighborsClassifier

training_data = [[1, 29, 1],
                 [2, 29, 0],
                 [3, 36, 1],
                 [1, 47, 0],
                 [2, 38, 1],
                 [3, 19, 1],
                 [1, 71, 0],
                 [2, 45, 0],
                 [3, 30, 1]]
training_target = [1, 0, 1, 1, 0, 1, 0, 0, 1]

knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(training_data, training_target)

test_data = [[2, 25, 0]]

prediction = knn.predict(test_data)

print("Predicted survival outcome for test data:", prediction)

if(prediction == 0):
  print("dead")
else:
  print("survived")
```

Step : 4 Now create and print a test_data array with the following cases from first to last in this order. 



```python
test_data = [[1, 4, 0], [1, 90, 0], [2, 43, 1], [2, 40, 0], [3, 17, 1], [3, 29, 0]]
predictions = knn.predict(test_data)
for i in range(len(test_data)):
    if predictions[i] == 1:
        print(f"Case {i+1}: {test_data[i]} - Predicted to survive")
    else:
        print(f"Case {i+1}: {test_data[i]} - Predicted to not survive")
```
